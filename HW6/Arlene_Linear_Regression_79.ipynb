### Part 1: Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

##### Extract features and label for the data

```python
data = pd.read_csv('E:\MS\MS_studies\Sem_2\IntrotoML\HW3\\71-80.csv')
```

```python
data = data.iloc[:, [3,4,5,6,7,8]]
```

##### Normalize the features[ Both X and Y]

```python
scaler = MinMaxScaler()

scaled_arr = scaler.fit_transform(data)
```

##### Split the data into training and testing sets

```python
trained_data, test_data = train_test_split(data,test_size=0.2, random_state=0)
```

```python
scaler = MinMaxScaler()

scaled_arr = scaler.fit_transform(trained_data)
```

```python
trained_data = pd.DataFrame(scaled_arr, columns=data.columns[0:6])
```

```python
trained_data
```

```python
X_train_scaled =  trained_data.iloc[:, [0,1,2,3,4]]
y_train_scaled = trained_data.iloc[:,[5]]
```

```python
X_train_scaled
```

##### Fit the linear model

```python
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

##### Test the test set

```python
scaled_arr = scaler.fit_transform(test_data)
```

```python
test_data = pd.DataFrame(scaled_arr, columns=data.columns[0:6])
```

##### Calculate MSE

```python
# Calculate squared differences
squared_errors = (y_actual - y_pred) ** 2
    
    # Calculate the mean of squared errors
mse = np.mean(squared_errors)
# Print the MSE
print(f"Mean Squared Error (MSE): {squared_errors}")
```

```python
mse
```

##### Calculate  Mean Absolute Error 

```python
mae = np.mean(np.abs(y_actual - y_pred))
```

```python
res_df = pd.DataFrame({'y_act' : y_actual, 'y_pred' : y_pred})
# res_df['absolute_error'] = abs(res_df['y_act'] - res_df['y_pred'])
```

```python
res_df['absolute_error'] = abs(res_df['y_act'] - res_df['y_pred'])
```

```python
res_df.absolute_error.mean()
```

```python
print(f"MAE for the above dataset using Linear Regression Model is {mae}")
```

##### Calculate Mean Absolute Percentage Error

```python
mape = np.mean(np.abs((y_pred.flatten() - y_actual.flatten())/y_actual.flatten())) * 100
```

```python
print(f"MAPE for the above dataset using Linear Regression Model is {mape}")
```
